```python
from qdrant_client import QdrantClient

client = QdrantClient(url="http://localhost:8029")
```

```python
from qdrant_client.models import Distance, VectorParams

client.create_collection(
    collection_name="test_collection",
    vectors_config=VectorParams(size=4, distance=Distance.DOT),
)
```

```python
from qdrant_client.models import PointStruct

import uuid
import random

points = [
    PointStruct(
        id=str(uuid.uuid4()),  # hoặc int(time.time() * 1000) + i
        vector=[random.uniform(0, 1) for _ in range(4)],
        payload={"city": city, "run": "lần 1"}
    )
    for i, city in enumerate(["Berlin", "London", "Moscow", "New York", "Beijing", "Mumbai"])
]

operation_info = client.upsert(
    collection_name="test_collection",
    wait=True,
    points=points
)
print(operation_info)
```
